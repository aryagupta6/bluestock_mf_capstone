## 1. Market Trends: NAV Dynamics (2022 – 2026)

**Chart: Daily NAV Trends for All 40 Schemes**

![NAV Trends](../reports/figures/01_nav_trends.png)

### Key Observations

- **Equity funds** exhibited high **volatility clustering**: a sharp rally during **2023**, followed by a series of volatile corrections in **mid-2024**.
- **Debt funds** maintained **steady, near-linear growth** with only minor corrections — a classic low-risk asset-class behaviour.
- COVID recovery momentum (early 2022) was clearly visible across all equity schemes, with NAVs accelerating upward.
- The divergence between Equity and Debt NAV trajectories widens after mid-2023, signalling increased risk appetite among fund managers.

**Insight:** The NAV volatility patterns confirm that equity mutual funds carry significantly higher short-term risk but higher long-term growth potential. Investors with shorter horizons should consider debt-heavy allocations.

## 2. AUM Growth by Asset Management Company (2022 – 2025)

**Chart: AUM Growth by AMC — Grouped Bar Chart**

![AUM Growth](../reports/figures/02_aum_growth.png)

### Key Observations

- **SBI Mutual Fund** holds the dominant position across all years, with AUM peaking at approximately **₹12.5 Lakh Crore** by end-2025.
- The top 3 AMCs (SBI, HDFC, ICICI Prudential) control a **disproportionately large share** of industry-wide AUM, highlighting market concentration.
- Year-over-year growth rates show acceleration post-2023, benefiting from the SIP-driven retail investment boom.

**Insight:** Market concentration risk exists — if a single large AMC underperforms, it could disproportionately affect millions of retail investors. Regulatory oversight on large AMCs is critical.

## 3. SIP Inflows: The Retail Revolution

**Chart: Monthly SIP Inflow Trend (Jan 2022 – Dec 2025)**

![SIP Inflows](../reports/figures/03_sip_inflows.png)

### Key Observations

- SIP inflows grew **exponentially** from **₹11,517 Crore** (Jan 2022) to a record high of **₹31,002 Crore** (Dec 2025) — a **~169% increase** over 4 years.
- The growth trajectory is consistently upward with only minor month-on-month dips, indicating sustained retail confidence.
- The ₹31,002 Crore milestone (Dec 2025) marks a new all-time record for monthly SIP contributions.

**Insight:** SIPs have become the primary entry vehicle for retail investors into equity markets. The consistent upward trend suggests strong financial literacy improvements and trust in mutual fund products.

## 4. Category-Level Inflow Patterns

**Chart: Net Mutual Fund Category Inflows — Heatmap**

![Category Inflow Heatmap](../reports/figures/04_category_inflow_heatmap.png)

### Key Observations

- **Equity categories** consistently attract the highest net inflows, with Large Cap and Flexi Cap leading.
- **Debt categories** show mixed patterns — Liquid and Ultra-Short Duration funds exhibit seasonal inflow/outflow cycles (likely treasury-management-driven).
- Some months show **sharp red zones** (net outflows) in specific categories, indicating tactical redemptions during market corrections.

**Insight:** Category-level heatmaps reveal that investor behaviour is not uniform — tactical shifts occur during volatile months, suggesting that fund houses should prepare for liquidity demands during correction periods.

## 5. Investor Demographics: Age Distribution

**Chart: Investor Age Group Distribution**

![Age Distribution](../reports/figures/05_age_distribution.png)

### Key Observations

- The **26–35 age group** dominates with **41.1%** of all transactions — the largest investor cohort.
- The **36–45 age group** follows at **24.9%**, together making the 26–45 demographic responsible for **~66%** of all mutual fund transactions.
- Seniors (56+) contribute only **7.6%** of transaction volume, suggesting lower digital adoption or preference for other asset classes.

**Insight:** Product design and marketing should focus heavily on the 26–45 demographic. However, the low 56+ participation represents an untapped market — simplified UX and offline support could help capture this segment.

## 6. SIP Transaction Amounts by Age Group

**Chart: Box Plot of SIP Transaction Amounts**

![SIP Box Plot](../reports/figures/06_sip_box_plot.png)

### Key Observations

- Although the **26–35 demographic** has the highest transaction frequency, the **46–55 age group** exhibits the **highest median SIP transaction amount**.
- This reflects higher disposable incomes and larger per-transaction investments in older brackets.
- The 18–25 group shows the lowest median amounts but a **wide interquartile range**, indicating diverse investment capacities among young investors.

**Insight:** Frequency ≠ Volume. Older investors invest less frequently but in larger amounts. Fund advisors should tailor SIP plans — smaller ticket sizes for younger investors, larger flexible amounts for the 46+ demographic.

## 7. Geographic Distribution of SIP Investments

**Chart: Total SIP Investment Amount by State**

![SIP by State](../reports/figures/07_sip_by_state.png)

### Key Observations

- **Punjab and Tamil Nadu** contributed the largest SIP transaction amounts, followed by **Madhya Pradesh** and **Rajasthan**.
- This challenges the assumption that mutual fund participation is concentrated in financial hubs (Mumbai, Delhi, Bengaluru).
- Deep capital market penetration is evident in Tier-2 and Tier-3 states.

**Insight:** Geographic diversification of mutual fund participation is a positive indicator for market depth. Fund houses should strengthen distribution networks in high-growth states outside the traditional metro corridor.

## 9. Industry Folio Count Growth

## 10. Return Correlation Between Funds

**Chart: Daily Return Pairwise Correlation Heatmap**

![NAV Correlation Matrix](../reports/figures/10_nav_correlation_matrix.png)

### Key Observations

- Equity funds within the same category (e.g., Large Cap) show **high positive correlation** (>0.85), as expected.
- **Cross-category correlations** (e.g., Large Cap vs Debt) are significantly lower, confirming diversification benefits.
- Some fund pairs exhibit **near-zero or negative correlation**, making them ideal for portfolio construction.

**Insight:** The correlation matrix provides a quantitative basis for portfolio diversification. Investors holding only within a single category are exposed to concentration risk. Multi-category allocation is strongly recommended.

## 11. Sector Allocation in Equity Holdings

**Chart: Sector Distribution of Stock Holdings**

![Sector Allocation](../reports/figures/11_sector_allocation.png)

### Key Observations

- The **Banking sector** constitutes the highest portfolio weight at **24.6%** of total equity holdings.
- **IT (17.2%)** and **Pharma (15.4%)** follow, with these three sectors together commanding **over 57%** of total equity allocation.
- This heavy concentration in three sectors poses systematic risk during sector-specific downturns.

**Insight:** Indian equity mutual funds are heavily overweight in Banking, IT, and Pharma. Fund managers should consider increasing allocation to underrepresented sectors (Energy, Infrastructure, Consumer) for better risk-adjusted returns.

## 12. Risk-Return Profile: Sharpe Ratio Distribution

**Chart: Distribution of Sharpe Ratios Across Schemes**

![Sharpe Ratio Distribution](../reports/figures/12_sharpe_ratio_distribution.png)

### Key Observations

- Sharpe ratios across the 40 schemes range from slightly negative to moderately positive.
- The majority of funds cluster around **0.3 – 0.8**, indicating moderate risk-adjusted performance.
- A few outliers with negative Sharpe ratios warrant closer inspection — these may be underperforming funds.

**Insight:** The Sharpe ratio distribution suggests that most funds deliver adequate risk-adjusted returns, but investors should avoid funds with consistently negative Sharpe ratios. These are destroying value relative to the risk-free rate.

## 13. Expense Ratio vs Performance Rating

**Chart: Expense Ratio vs Morningstar Star Rating**

![Expense vs Rating](../reports/figures/13_expense_vs_rating.png)

### Key Observations

- **Direct plans** have significantly lower expense ratios (averaging below **0.9%**) compared to regular plans.
- Despite lower costs, direct plans maintain **similar or higher Morningstar star ratings**.
- There is a **weak negative correlation** between expense ratio and performance rating — cheaper funds tend to rate slightly better.

**Insight:** This is a strong endorsement for Direct Plan investing. Investors paying higher expense ratios through regular plans are not receiving proportionally better performance. The data supports the migration towards low-cost direct plans.

## 14. Transaction Behaviour by Payment Mode

**Chart: Average Transaction Amount by Payment Mode**

![Average Transaction by Mode](../reports/figures/14_avg_transaction_by_mode.png)

### Key Observations

- Different payment modes show varying average transaction sizes.
- **Bank transfer/NEFT** transactions tend to have higher average amounts, typical of larger lump-sum investments.
- **UPI/digital** payments show lower average amounts but higher frequency — consistent with SIP behaviour.

**Insight:** Payment mode choice is a proxy for investor type — larger institutional/HNI investors prefer bank transfers, while retail SIP investors prefer UPI. Platform design should optimise the UPI checkout experience to capture the growing retail segment.

## 15. Fund Quality: Morningstar Ratings Distribution

**Chart: Morningstar Star Ratings Distribution**

![Morningstar Ratings](../reports/figures/15_morningstar_rating_distribution.png)

### Key Observations

- The distribution of Morningstar ratings across the 40 schemes shows a **right-skewed** pattern — most funds are rated 3 stars or above.
- Very few funds fall below 2 stars, indicating that the selected fund universe is generally well-managed.
- 4-star and 5-star funds are in the minority, confirming that elite performance is rare and difficult to sustain.

**Insight:** Morningstar ratings serve as a useful first-pass filter for fund selection. However, since most funds cluster at 3–4 stars, investors should combine star ratings with other metrics (Sharpe ratio, expense ratio, category ranking) for better decision-making.

## Conclusion & Recommendations

### For Fund Houses & AMCs:
1. **Diversify sector allocation** — reduce over-reliance on Banking, IT, and Pharma to improve risk-adjusted returns.
2. **Expand B30 distribution** — digital-first strategies and vernacular language support can accelerate adoption in smaller cities.
3. **Target women investors** — dedicated financial literacy programs and simplified onboarding can close the gender gap.

### For Investors:
1. **Prefer Direct Plans** — lower expense ratios without sacrificing performance quality.
2. **Diversify across categories** — the correlation matrix shows clear diversification benefits from multi-category allocation.
3. **Avoid funds with negative Sharpe ratios** — these are underperforming relative to the risk-free rate.

### For Regulators:
1. **Monitor AMC concentration** — the top 3 AMCs control a disproportionate market share.
2. **Promote financial inclusion** — the B30/T30 gap and gender disparity need structural intervention.
3. **Ensure SIP sustainability** — with ₹31,000+ Crore monthly inflows, systemic risk management becomes critical.
